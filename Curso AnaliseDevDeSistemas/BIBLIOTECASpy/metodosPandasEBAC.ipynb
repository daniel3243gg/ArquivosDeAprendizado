```python
import pandas as pd
#metodo que soma uma coluna
github_df = pd.read_csv('./github.csv', sep=';')
sum_series = github_df[['stars', 'stars_today']].agg('sum')
sum_series
```

```python
import pandas as pd
#agrupando colunas
linguagem_metricas = github_df[['language', 'stars', 'stars_today']].groupby('language').agg('sum')
linguagem_metricas
```

```python
import pandas as pd

#agrupandao varias metricas
agrupamento_de_varias_metricas = github_df[['language', 'stars', 'stars_today']].groupby('language').agg(['count', 'sum', 'mean', 'std'])
agrupamento_de_varias_metricas
```
