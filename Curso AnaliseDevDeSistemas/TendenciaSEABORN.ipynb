```python
import seaborn as sns
data = sns.load_dataset("flights")
data.head()
```

```python
#ordenando
flights = data[["year", "passengers"]]
flights = flights.groupby("year").agg("sum").reset_index()
#CRIANDO O GRAFICO TEMPORAL
with sns.axes_style('whitegrid'): #escolhe a gride por tras do grafico.
  grafico = sns.lineplot(data=flights, x="year",  y="passengers", palette="pastel")
  grafico.set(title='Passageiros por ano', xlabel='Ano', ylabel='Passageiros');
```

```python
#TENDENCIA AREA
flights = data[["year", "passengers"]]
flights = flights.groupby("year").agg("sum").reset_index()
flights.head()
import matplotlib.pyplot as plt
with sns.axes_style('whitegrid'):
  grafico = sns.FacetGrid(data=flights, palette="pastel") #criando o grafico
  grafico.map(sns.lineplot, "year", "passengers") #criando a linha
  grafico.map(plt.fill_between, 'year', 'passengers', alpha=0.3) #pintando em baixo
  grafico.set(title='Passageiros por ano', xlabel='Ano', ylabel='Passageiros'); #dando os title
  grafico.fig.set_size_inches(w=15/2.54,  h=7.5/2.54)#definindo o tamanho
```
