```python
#grafico de dispersao
import seaborn as sns
data = sns.load_dataset("iris")
data.head()

iris = data[["sepal_length", "petal_length", "species"]]
with sns.axes_style('whitegrid'):
  grafico = sns.scatterplot(data=iris, x="petal_length", y="sepal_length", hue="species", palette="pastel")
  grafico.set(title='Comprimento da ptala por comprimento da spala', xlabel='Ptala (cm)', ylabel='Spala (cm)');
  grafico.get_legend().set_title("Espcie");
```

```python
#GRAFICO DE CALOR

data = sns.load_dataset("flights")
data.head()
flights = data.pivot("month", "year", "passengers") #PIVO GERA UM NOVO DATAFRAME COLOCANDO MONTH COMO LINHA  YEAR COMO COLUNA(TIPO XADREZ)
flights.head()
grafico = sns.heatmap(data=flights, cmap="Spectral")
grafico.set(title='Passageiros por ms por ano', xlabel='Ano',  ylabel='Ms');

grafico = sns.heatmap(data=flights, cmap="Spectral", annot=True,  fmt="d") #ANNOT mostra os valores no grafico
grafico.set(title='Passageiros por ms por ano', xlabel='Ano', ylabel='Ms');
grafico.figure.set_size_inches(w=20/2.54,  h=10/2.54)
```
