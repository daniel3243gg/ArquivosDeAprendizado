```python
import seaborn as sns

sns.get_dataset_names() #DATAFRAMES PRONTOS do pacote seaborn
```

```python
data = sns.load_dataset("tips")
data.head(10)
```

```python
grafico = sns.relplot(data=data,x="total_bill", y="tip", hue="day", style="day")#DATA = nome do dataframe, Hue= definir a legenda o significado de cada bolinha, Style = é o estilo.
sns.relplot(data=data,x="total_bill", y="tip", hue="day", style="day", col="time")#col = cria outra variavel de grafico
sns.relplot(data=data,x="total_bill", y="tip", hue="day", style="day",  col="time", row="sex")#ROW= cria graficos divididos em colunas
```

```python
grafico.set_titles('Gorjetas', fontsize=20,fontweight='bold' )#definindo o titulo do grafico
grafico.set_xlabels('VALOR DA CONTA(USD)', fontsize=12)#definindo um nome para o eixo X
grafico.set_ylabels('VALOR DA GORJETA(USD)', fontsize=12)#DEFININDO um nome para o eixo Y
grafico.legend.set_title('DIAS DA SEMANA', prop={'size':10, 'weight':'bold'})#definindo a legenda
grafico.fig
```

```python
#paleta de cores pode ser difinido atraves do argumento palette = 'PALETA DESEJADA'
grafico = sns.relplot(data=data,x="total_bill", y="tip", hue="day", style="day", palette='dark')
grafico.set_xlabels('VALOR DA CONTA(USD)', fontsize=12)#definindo um nome para o eixo X
grafico.set_ylabels('VALOR DA GORJETA(USD)', fontsize=12)#DEFININDO um nome para o eixo Y
grafico.legend.set_title('DIAS DA SEMANA', prop={'size':10, 'weight':'bold'})#definindo a legenda
grafico.fig
```

```python
#LIDANDO COM O TAMANHO DO GRAFICO

altura= 20 / 2.54 #2.54 é o padrao de polegadas
largura= 20 / 2.54

grafico.fig.set_size_inches(w=altura,h=largura)#mudando o tamanho da figura
grafico.fig
```

```python
#salvando a figura
grafico.fig.savefig(fname="gorjetas.png",  bbox_inches="tight") #o argumetno bbox_inches é o padrao para a imagem nao cortar.
grafico.fig.savefig(fname="gorjetas.pdf",  bbox_inches="tight") #existe inumero de salvar alem de png é pdf
```
