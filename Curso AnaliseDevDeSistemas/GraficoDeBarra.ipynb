```python
import seaborn as sns
data = sns.load_dataset("tips")
data.head()

tips = data[["day", "total_bill"]].groupby("day").agg("sum").reset_index() #agrupando os dias da semana

grafico= sns.barplot(data=tips, x='day', y='total_bill', ci=None, palette = 'pastel')#cRIANDO grafico de barras
grafico.set(title='TOTAL DA SEMANA', xlabel='DIAS DA SEMANA', ylabel= 'TOTAL')#definindo os titles de umz vez so

```

```python
#invertendo o grafico para facilitar a vizualizaçao

grafico= sns.barplot(data=tips, x='total_bill', y='day', ci=None, palette = 'pastel')#perceba que tomos o x/y de lugares
```

```python
#adicionando mais uma variavavel ao grafico
tips_1 = data[['day', 'total_bill', 'time']].groupby(['day', 'time']).agg('sum').reset_index()#ordenando para adicionar mais uma variavel no grafico
tips_1
grafico =sns.barplot(data=tips_1, x='total_bill', y='day',hue='time', ci=None, palette = 'pastel')
```
